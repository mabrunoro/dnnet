### Exe. 2
Escreva um programa para calcular a frequência das palavras a partir de uma string de entrada. 
A saída deve ser exibida após a classificação alfanumérica da chave.

Exemplo:
Suponha que o seguinte string de entrada seja fornecida ao programa:
```python
str_input = 'Novo para Python ou escolher entre Python 2 e Python 3? Leia Python 2 ou Python 3'
```
Em seguida, a saída do programa deve ser:
```
2:2
3:1
3?:1
Leia:1
Novo:1
Python:5
e:1
entre:1
escolher:1
ou:2
para:1
```
Sugestões: use um dicionário para guardar os pares (palavra, frequência) .

````python
#inserir tua solução aqui
````

````python
freq = {}   # frequency of words in text
line = 'Novo para Python ou escolher entre Python 2 e Python 3? Leia Python 2 ou Python 3'
for word in line.split():
    freq[word] = freq.get(word,0)+1

words = list(freq.keys())
words.sort()

for w in words:
    print("%s:%d" % (w,freq[w]))
````

## Funções <a name="Func"></a>

### Procedimento

````python
def seqFibonacci1(n):
  #Escreve a serie de Fibonacci ate n
	a,b = 0,1
	print(a)
	while(b<n):
		print(b)
		a,b = b, a+b
    
seqFibonacci1(7)
````

### Função

````python
def seqFibonacci2(n):
  #Escreve a serie de Fibonacci ate n
  a,b = 0,1
  seq = [a]
  while(b<n):
    a,b = b, a+b
    seq.append(a)
  return seq

print(seqFibonacci2(7))
````

### Sentença for

````python
#Percorrendo a lista, obtendo os valores e seus índices:
for indice, valor in enumerate(lista):
    print("lista[%d] = %d" % (indice, valor))
````

````python
# Percorrendo um pedaço de uma lista usando slicing:
for x in lista[1:4]:
    print(x)
````

````python
# Percorrendo uma lista de trás pra frente 
# definindo o passo do slicing como -1:
for x in lista[::-1]:
    print(x)
````
