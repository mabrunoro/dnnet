## Indexação <a name="Indexation"></a>

### Introdução

```python
a = np.arange(5)
print('a = ',a)
a[0] = 1 
a[-1] = a[0] + a[2]
print('a = ',a)
```

```python
a = np.diag(np.arange(2))
print('a = ',a)
a[0, 1] = 12 # primeira linha, segunda coluna
a[1, 1] = 10 # segunda linha, segunda coluna
print('a = ',a)
```

### Indexação sofisticada

#### Usando máscaras booleanas

```python
np.random.seed(3)
a = np.random.random_integers(0, 20, 10)
print('a = ',a)
```

```python
mask = (a % 3 == 0) # Criando a mascara
print('mask = ',mask)
```

```python
subarray = a[mask] # extraindo um subarray de a com mask
print('subarray = ', subarray)
```

```python
a[mask] = -1 # atribuindo um novo valor a elementos de a com mask
print('a = ',a)
```

#### Usando um arrays de inteiros

```python
a = np.arange(10)
print('a = ',a)
# Criando a lista de índices, note que idx é uma lista do Python
idx = [2, 3, 9, 4, 2]
b = a[idx] # extraindo um subarray de a com idx 
print('b = ',b)
a[idx] = -10 # atribuindo um novo valor a elementos de a com idx
print('a = ',a)
```

## Fatiamento em arrays unidimensionais (1D) <a name="FatArray1D"></a>

### Exemplo simples de fatiamento

```python
a = np.arange(20) # a é um vetor de dimensão 20
print('a = ', a)
print('a[1:15:2] = ',a[1:15:2])
```

### Exemplo de fatiamento com indices negativos

#### Acessando o último elemento com índice negativo

```python
a = np.arange(20) # a é um vetor de dimensão 20
print('a = ', a)
# Note que o fatiamento termina antes do último elemento (-1)
print('a[1:-1:2] = ',a[1:-1:2])
```

#### Inversão do array com step negativo (step = -1)

```python
a = np.arange(20) # a é um vetor de dimensão 20
print('a = ', a)
# Note que o fatiamento retorna o array invertido
# Antepenúltimo até o terceiro elemento com step = -1
print('a[-3:2:-1] = ',a[-3:2:-1])
```

### Fatiamento avançado

#### Supressão do índice limite inferior

```python
a = np.arange(20) # a é um vetor de dimensão 20
print('a = ', a)
# Note que o fatiamento inicia do primeiro elemento
# Primeiro elemento até antes do 15o com passo duplo 
print('a[:15:2] = ', a[:15:2])
```

#### Supressão do índice limite superior

```python
a = np.arange(20) # a é um vetor de dimensão 20
print('a = ', a)
# Note que o fatiamento termina no último elemento
# Primeiro elemento até o último com passo duplo
print('a[1::2] = ', a[1::2])
```

#### Supressão do índice do passo

```python
a = np.arange(20) # a é um vetor de dimensão 20
print('a = ', a)
# Note que o fatiamento tem passo unitário
# Primeiro elemento até antes do 15o com passo um
print('a[1:15] = ', a[1:15])
```

#### Todos os elementos com passo unitário

```python
a = np.arange(20) # a é um vetor de dimensão 20
print('a = ', a)
print('a[:] = ',a[:])
```

## Fatiamento em arrays bidimensionais (2D) <a name="FatArray2D"></a>

### Inicializando um array e mudando o seu shape

```python
a = np.arange(20) # a é um vetor de dimensão 20
print('a = ', a)
a = a.reshape(4,5) # a agora é um array 4x5 (4 linhas por 5 colunas)
print('a.reshape(4,5) = \n', a)
```

### Fatiamento de linhas e colunas de um array

```python
# A segunda linha do array
print('a[1,:] = ', a[1,:])
# A primeira coluna do array 
print('a[:,0] = ', a[:,0])
#Acessando as linhas do array de 2 em 2 começando pelo índice 0 
print('a[0::2,:] = ', a[0::2,:])
#Acessando as linhas e colunas do array de 2 em 2 começando pela #linha 0 e coluna 1 
print('a[0::2,1::2] = ', a[0::2,1::2])
#Acesso as duas últimas linhas do array em ordem reversa
b = a[-1:-3:-1,:]
print('a[-1:-3:-1,:] = ',b)
#Invertendo a ordem das linhas do array:
c = a[::-1,:]
print('a[::-1,:] = ',c)
```

## Copiando variáveis `ndarray` <a name="cpoyVar"></a>

### Sem cópia explícita, apenas referência

```python
a = np.arange(6) 
b = a        #Sem cópia explícita, apenas referência
print("a =",a)
print("b =",b)
b.shape = (2,3)                 # mudança no shape de b,
print("a shape =",a.shape)      # altera o shape de a
b[0,0] = -1     # mudança no conteúdo de b
print("a =",a)  # altera o conteudo de a 
print("id de a = ",id(a)) # id é um identificador único de objeto
print("id de b = ",id(b)) # a e b possuem o mesmo id 
```

```python
def cc(a):
  return a
a = np.arange(6)
b = cc(a)
print("id de a = ",id(a))
print("id de b = ",id(b))
```

### Cópia profunda

```python
a = np.arange(16).reshape((4,4))
b = np.array(a)
print("id de a = ",id(a))
print("id de b = ",id(b))
```

## Operações numéricas com Arrays <a name="OpArray"></a>

### Operações Matriciais

```python
a = np.arange(20).reshape(5,4)
b = 2*np.ones((5,4))
c = np.arange(20,0,-1).reshape(4,5)
print('A = ',a)
print('B = ',b)
print('C = ',c)
# Multiplicação de matriz por escalar
print("B*5 = ", b*5)
# Soma de matrizes
print("A+B = ", a+b)
# Produto das matrizes A e B, elemento por elemento
print("A*B = ", a*b)
# Produto das matrizes A e C
print("A*C = ",np.dot(a,c))
# Transposta de A
print("AT = ", a.transpose())
```

### Operações de reduçã*o*

```python
A = np.arange(24).reshape(2, 3, 4)
print("A_sum_colunas = ", np.sum(A,axis=0))
print("A_sum_linhas  = ", np.sum(A,axis=1))
print("A_sum_deep    = ", np.sum(A,axis=2))
```

```python
# Computando a soma sobre os elementos da matriz A
A = np.arange(20).reshape(5,4)
print("A = ", A)
print("Sum_elements_A = ", np.sum(A))
print("A_sum_colunas = ", np.sum(A,axis=0))
print("A_sum_linhas  = ", np.sum(A,axis=1))
```

```python
# Computando os valores médios sobre os elementos da matriz A
A = np.arange(20).reshape(5,4)
print("A = ", A)
print("Mean_elements_A = ", np.mean(A))
print("A_mean_colunas = ", np.mean(A,axis=0))
print("A_mean_linhas  = ", np.mean(A,axis=1))
```

```python
# Computando os valores min. e max. sobre os elementos da matriz A
A = np.random.randint(20, size=(2, 4))
print("A = ", A)
print("Valor minimo de A = ", np.min(A))
print("Valor maximo de A = ", np.max(A))
print("indice do valor minimo de A = ", np.argmin(A))
print("indice do valor maximo de A = ", np.argmax(A))
```

### Broadcasting (expansão)

```python
a = np.array([[1, 2], [3, 4], [5, 6]], float)
b = np.array([-1, 3], float)
c = a + b
print('a = ',a)
print('b = ',b)
print('a + b = ',c) 
```

```python
a = np.zeros((2,2), float)
b = np.array([-1, 3], float)
print('a = ',a)
print('b = ',b)
#Expansão no eixo das linhas
c1 = a + b[np.newaxis, :]
print('a + b[np.newaxis, :] = ',c1)
#Expansão no eixo das colunas
c2 = a + b[:, np.newaxis]
print('a + b[:, np.newaxis] = ',c2)
```

## Funções interessantes de Numpy <a name="FunNumpy"></a>

### Função `tile`

```python
a = np.array([0, 1, 2])
b = np.tile(a,2)
c = np.tile(a,4)
d = np.tile(a,6)
print('a =',a)
print('b =',b)
print('c =',c)
print('d =',d)
```

```python
a = np.arange(4).reshape(2,2)
b = np.tile(a,2)
c = np.tile(a,4)
d = np.tile(a,6)
print('a =',a)
print('b =',b)
print('c =',c)
print('d =',d)
```

```python
a = np.arange(4).reshape(2,2)
b = np.tile(a,(3,1))
print('a =',a)
print('b =',b)
```

```python
a = np.arange(4).reshape(2,2)
b = np.tile(a,(2,2))
print('a =',a)
print('b =',b)
```

### Função `ravel`

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.ravel(a)
print('a =',a) 
print('b =',b)
```

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.ravel(a,'F')
print('a =',a) 
print('b =',b)
```

### Função reshape

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.ravel(a)
c1 = np.reshape(a,(3,2))
c2 = np.reshape(a,(2,3))
print('a =',a) 
print('b =',b)
print('c1 =',c1)
print('c2 =',c2)
```

### Função resize

```python
a = np.array([[0,1],[2,3]])
b = np.resize(a,(1,7))
c = np.resize(a,(2,5))
print('a =',a)
print('b =',b)
print('c =',c)
```

### Função clip

```python
a = np.array([11,1,2,3,4,5,12,-3,-4,7,4])
b = np.clip(a,0,10)# filtra valores<0 e valores>10 
print('a =',a)  
print('b =',b)
```

### Função unique

```python
a = np.array([1, 1, 4, 5, 5, 5, 7], float)
b = np.unique(a) 
print('a =',a)  
print('b =',b)
```
